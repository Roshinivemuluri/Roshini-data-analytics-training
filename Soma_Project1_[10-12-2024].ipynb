## Task-2: Create new columns:##
 ## * Day of the week.

```python
# prompt: Create new columns:
# o Day of the week.
# o Day of the month.
# o Month.
# o Year.
# o Season (Winter, Spring, Summer, Fall).

# Extract date components
data['day_of_week'] = data['transaction_date'].dt.dayofweek
data['day_of_month'] = data['transaction_date'].dt.day
data['month'] = data['transaction_date'].dt.month
data['year'] = data['transaction_date'].dt.year

# Define a function to determine the season
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'

# Apply the function to create the 'season' column
data['season'] = data['month'].apply(get_season)

data.head()
```

##Summary of Heatmap Visualization##

A heatmap is a graphical representation of data where values are depicted by color. It is essentially a two-dimensional grid where each cell's color intensity corresponds to the magnitude of the value it represents.

Key Features

Data Representation: Heatmaps use color to represent numerical data, making patterns and relationships easily visible.
Grid Structure: Data is displayed in a grid format, with rows and columns representing different variables or categories.
Color Intensity: The intensity or shade of color in each cell indicates the magnitude of the value. Typically, darker or warmer colors represent higher values, while lighter or cooler colors represent lower values.
Applications: Heatmaps are commonly used to visualize correlations between variables, identify clusters or patterns in data, and explore relationships in large datasets.
Benefits

Easy Interpretation: Heatmaps provide a visually intuitive way to understand complex data patterns and relationships.
Pattern Identification: They help identify clusters, outliers, and trends in data that might not be apparent in other visualizations.
Correlation Visualization: Heatmaps are particularly effective in visualizing correlations between multiple variables.
Data Exploration: They enable users to explore large datasets and quickly identify areas of interest.

##Summary of Observations about Key Correlations##

After generating a correlation heatmap, you can observe the relationships between different variables in your dataset. Here's a brief summary of how to interpret key correlations:

Strong Positive Correlation: If two variables have a correlation coefficient close to +1, it indicates a strong positive linear relationship. As one variable increases, the other tends to increase as well. This is often represented by dark red or warm colors on the heatmap.

Strong Negative Correlation: If two variables have a correlation coefficient close to -1, it indicates a strong negative linear relationship. As one variable increases, the other tends to decrease. This is often represented by dark blue or cool colors on the heatmap.

Weak or No Correlation: If two variables have a correlation coefficient close to 0, it suggests a weak or no linear relationship between them. Changes in one variable do not have a predictable effect on the other. This is often represented by light colors or white on the heatmap.

##Explanation of when to use loc vs iloc##
Use it for label-based indexing: When you want to access data using row and column labels (names).
Syntax: df.loc[row_label, column_label]
Examples:
df.loc['row_name', 'column_name']: Access a single value.
df.loc[['row1', 'row2'], ['col1', 'col2']]: Access a subset of rows and columns.
df.loc['row1':'row5', :]: Access a range of rows by label.
df.loc[df['column_name'] > 10, :]: Access rows based on a condition.
iloc

Use it for integer-based indexing: When you want to access data using row and column positions (numbers).
Syntax: df.iloc[row_position, column_position]
Examples:
df.iloc[0, 0]: Access the value in the first row and first column.
df.iloc[[0, 2], [1, 3]]: Access specific rows and columns by position.
df.iloc[:5, :]: Access the first 5 rows.
df.iloc[df['column_name'].argsort()[:10], :]: Access the top 10 rows based on the sorted values of a column.
In Summary:

loc: Use for labels (names).
iloc: Use for integers (positions).
Choosing between loc and iloc:

If you know the names of your rows and columns, use loc.
If you know the positions of your rows and columns, use iloc.
If you are working with a DatetimeIndex (like in your example), loc is generally used for selecting data based on date labels or slices, but it's important to be aware of the potential issues with non-monotonic indexes and missing dates. In such cases, filtering with boolean indexing might be a more robust approach.
