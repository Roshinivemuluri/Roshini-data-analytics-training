```
import statistics
from scipy.stats import skew

# Laptop dataset
laptops = [
    ["Apple", "MacBook Pro", "Ultrabook", 13.3, "IPS Panel Retina Display 2560x1600", "Intel", "Core i5", 2.3, 8, "128GB SSD", "Intel", "Iris Plus Graphics 640", "macOS", 1.37, 1339.69],
    ["Apple", "Macbook Air", "Ultrabook", 13.3, "1440x900", "Intel", "Core i5", 1.8, 8, "128GB Flash Storage", "Intel", "HD Graphics 6000", "macOS", 1.34, 898.94],
    ["HP", "250 G6", "Notebook", 15.6, "Full HD 1920x1080", "Intel", "Core i5 7200U", 2.5, 8, "256GB SSD", "Intel", "HD Graphics 620", "No OS", 1.86, 575.00],
    ["Apple", "MacBook Pro", "Ultrabook", 15.4, "IPS Panel Retina Display 2880x1800", "Intel", "Core i7", 2.7, 16, "512GB SSD", "AMD", "Radeon Pro 455", "macOS", 1.83, 2537.45],
    ["Apple", "MacBook Pro", "Ultrabook", 13.3, "IPS Panel Retina Display 2560x1600", "Intel", "Core i5", 3.1, 8, "256GB SSD", "Intel", "Iris Plus Graphics 650", "macOS", 1.37, 1803.60],
]

# Extracting Continuous Numeric Data for analysis
screen_sizes = [row[3] for row in laptops]  # Screen sizes
weights = [row[13] for row in laptops]      # Weights
prices = [row[14] for row in laptops]       # Prices

print("Screen Sizes:", screen_sizes)
print("Weights:", weights)
print("Prices:", prices)

# ---- Statistical Analysis ----

# Mean
mean_screen_size = statistics.mean(screen_sizes)
mean_weight = statistics.mean(weights)
mean_price = statistics.mean(prices)

# Median
median_screen_size = statistics.median(screen_sizes)
median_weight = statistics.median(weights)
median_price = statistics.median(prices)

# Mode (Screen sizes can have a mode)
try:
    mode_screen_size = statistics.mode(screen_sizes)
except statistics.StatisticsError:
    mode_screen_size = "No unique mode"

# Range
range_screen_size = max(screen_sizes) - min(screen_sizes)
range_weight = max(weights) - min(weights)
range_price = max(prices) - min(prices)

# Variance
variance_screen_size = statistics.variance(screen_sizes)
variance_weight = statistics.variance(weights)
variance_price = statistics.variance(prices)

# Standard Deviation
std_dev_screen_size = statistics.stdev(screen_sizes)
std_dev_weight = statistics.stdev(weights)
std_dev_price = statistics.stdev(prices)

# Skewness
skewness_screen_size = skew(screen_sizes)
skewness_weight = skew(weights)
skewness_price = skew(prices)

# ---- Results ----
print("\nStatistical Analysis Results:")
print(f"Mean (Screen Size): {mean_screen_size}")
print(f"Mean (Weight): {mean_weight}")
print(f"Mean (Price): {mean_price}\n")

print(f"Median (Screen Size): {median_screen_size}")
print(f"Median (Weight): {median_weight}")
print(f"Median (Price): {median_price}\n")

print(f"Mode (Screen Size): {mode_screen_size}\n")

print(f"Range (Screen Size): {range_screen_size}")
print(f"Range (Weight): {range_weight}")
print(f"Range (Price): {range_price}\n")

print(f"Variance (Screen Size): {variance_screen_size}")
print(f"Variance (Weight): {variance_weight}")
print(f"Variance (Price): {variance_price}\n")

print(f"Standard Deviation (Screen Size): {std_dev_screen_size}")
print(f"Standard Deviation (Weight): {std_dev_weight}")
print(f"Standard Deviation (Price): {std_dev_price}\n")

print(f"Skewness (Screen Size): {skewness_screen_size}")
print(f"Skewness (Weight): {skewness_weight}")
print(f"Skewness (Price): {skewness_price}")
```
