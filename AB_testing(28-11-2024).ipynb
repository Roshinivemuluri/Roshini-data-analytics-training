```python
import pandas as pd
from scipy import stats
```

```python
data = pd.read_csv('/content/ecommerce-ab-testing/ab_test.csv')
```

```python
data
```

```python
control_group = data[data['con_treat'] == 'control']['converted']
treatment_group = data[data['con_treat'] == 'treatment']['converted']
```

```python
t_statistic, p_value = stats.ttest_ind(control_group, treatment_group)
```

```python
alpha = 0.05  # significance level

print(f"T-statistic: {t_statistic}")
print(f"P-value: {p_value}")

if p_value < alpha:
    print("There is a statistically significant difference in conversion rates between the two groups.")
else:
    print("There is no statistically significant difference in conversion rates between the two groups.")
```
